```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
Base = pd.read_excel("Datos.xlsx", sheet_name = 'CLIENTE1')
Base['Numero_Cliente'] = 1

for Numero_Cliente in list(range(2,21)):
    Base_Temp = pd.read_excel("Datos.xlsx", sheet_name = 'CLIENTE'+ str(Numero_Cliente))
    Base_Temp['Numero_Cliente'] = int(Numero_Cliente)
    Base = pd.concat([Base, Base_Temp], ignore_index=True)
```

```python
for Variable in ['Presion','Temperatura','Volumen']:
    print(Variable)
    Estadisticas = Base.groupby('Numero_Cliente')[Variable].agg(Cantidad='count',
                                                                 Media='mean',
                                                                 Desviación_Estándar='std',
                                                                 Min='min',
                                                                 Q1=lambda x: x.quantile(0.25),
                                                                 Mediana='median',
                                                                 Q3=lambda x: x.quantile(0.75),
                                                                 Max='max',
                                                                 Nulos=lambda x: x.isnull().sum()).reset_index()


    Estadisticas = Estadisticas.sort_values(by='Numero_Cliente').reset_index(drop=True).round(2)
    display(Estadisticas)
```

```python
Base['Año'] = Base['Fecha'].dt.year
Base['Mes'] = Base['Fecha'].dt.month

Volumen_Estacionalidad = Base.groupby(['Año', 'Mes'])['Volumen'].sum().reset_index()
Volumen_Estacionalidad['Volumen'] = Volumen_Estacionalidad['Volumen'] / 1000

plt.figure(figsize=(15, 8))
Años_Unicos = Volumen_Estacionalidad['Año'].unique()
Colores = plt.cm.get_cmap('tab20', len(Años_Unicos)) 
Marcadores = ['o', 's', 'D', '^', 'v', 'p', '*', 'H', '<', '>']

for idx, Año in enumerate(Años_Unicos):
    Datos_Año = Volumen_Estacionalidad[Volumen_Estacionalidad['Año'] == Año]
    plt.plot(Datos_Año['Mes'], 
             Datos_Año['Volumen'], 
             label = Año, 
             color = Colores(idx), 
             marker = Marcadores[idx % len(Marcadores)])

plt.title('Estacionalidad del Consumo de Gas por Año (Volumen Total en Miles)', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Volumen Total (Miles)', fontsize=14)
plt.xticks(ticks=range(1, 13), labels=['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'], fontsize=12)
plt.legend(title='Años', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False) 
plt.tight_layout()
plt.show()
```

```python
Base['Periodo'] = Base['Fecha'].dt.to_period('M')
Volumen_Mensual = Base.groupby(['Periodo', 'Numero_Cliente'])['Volumen'].sum().reset_index()
Volumen_Mensual['Periodo'] = Volumen_Mensual['Periodo'].dt.to_timestamp()


Clientes_Unicos = Volumen_Mensual['Numero_Cliente'].unique()
colormap = cm.get_cmap('tab20', len(Clientes_Unicos))  # Paleta con hasta 20 colores distintos
Colores = {Cliente: colormap(idx) for idx, Cliente in enumerate(Clientes_Unicos)}
Marcadores = ['o', 's', 'D', '^', 'v', 'p', '*', 'H', '<', '>']

plt.figure(figsize=(15, 8))
for idx, Cliente in enumerate(Clientes_Unicos):
    Datos_cliente = Volumen_Mensual[Volumen_Mensual['Numero_Cliente'] == Cliente]
    plt.plot(Datos_cliente['Periodo'], 
             Datos_cliente['Volumen'], 
             label=f'Cliente {Cliente}', 
             color=Colores[Cliente], 
             marker=Marcadores[idx % len(Marcadores)])


plt.title('Consumo en Volumen por Mes y Cliente', fontsize=16)
plt.xlabel('Periodo', fontsize=14)
plt.ylabel('Volumen', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Clientes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.tight_layout()
plt.show()
```
